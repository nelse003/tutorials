### Defining functions

We are not limited to existing functions we can define our own:

````python
def show_favourite_chocolate(favourite: str, chocolates: list) -> bool:
    for sweet in chocolates:
        if sweet == favourite:
            print(f"My favourite - {sweet}!")
            return True
        else:
            print(f"Darn it's a - {sweet}")
    print("My favourite isn't there ")
    return False
````

Note where we add the : str and : list, we are hinting to the user of the code what types the argument of the function is expecting. We also hint what the function will return. A bool or Boolean is a True/ False. 

````python
show_favourite_chocolate(favourite="Wispa",chocolates=heroes)
````

````python
show_favourite_chocolate(favourite="Green Triangle",chocolates=heroes)
````

````python
show_favourite_chocolate(favourite="Wispa",chocolates=lots_heroes)
````

Note we are slightly breaking our do one thing principle with this function, it is both printing when the chocolate appears in the list and testing whether the sweet is in the list. We definitely could seperate this funcationality:

````python
def print_favourite_chocolate(favourite: str, chocolates: list) -> None:
    for sweet in chocolates:
        if sweet == favourite:
            print(f"My favourite - {sweet}!")
        else:
            print(f"Darn it's a - {sweet}")
    print("My favourite isn't there ")
````

## Data Structures - Dictionary - Too many eclairs!!

<img src="https://i.redd.it/bt0f2li81m561.jpg" width=600px>

When we modelled the box of heroes as 5 lots of each chocolate, we had a somewhat inaccurate model. Cadburys/ Kraft give us more eclairs than twirls. There are a few ways we could model this. Let us use a dictionary:

````python
# Dictionary of sweets with their relative weights
heroes_distribution = {
    "Fudge": 9,
    "Dairy Milk Caramel": 7,
    "Dairy Milk": 7,
    "Wispa": 5,
    "Twirl": 5,
    "Creme Egg Twisted": 8,
    "Eclair": 9,
    "Dinky Decker": 8,
    "Crunchie Bits": 7
}
````

A dictionary is a key value pair, in this case the chocolate name is the key and the number of chocolates in a tub is the value. 

````python
heroes_distribution['Fudge']
````

````python
heroes_distribution.keys()
````

````python
def generate_random_sweets_list(size: int, distribution: dict[str, float])-> list:
    return random.choices(list(distribution.keys()), weights=distribution.values(), k=size)
````

This function is just a single line, why would we write it as a function?

- It is more specific to the problem than using the random.choices function directly
- It has a name that is more obvious to what it does
- We can specify the types of the arguments to make it more clear what it does
- We can specify what it returns, so we know what object to expect
- We could change how it works (i.e. use a different random number generator), but the users of the function wouldn't need to change code that uses it

Why not?
- It adds more things to remember if someone already knows the random module well

````python
chocs = generate_random_sweets_list(100, distribution=heroes_distribution)
````

````python
print(chocs)
````

##  While loop - Actually eating the chocolates

Although for loops are generally much more common, while loops that continue until some condition is met can be useful in some cases:

````python
chocs = generate_random_sweets_list(15, distribution=heroes_distribution)
while len(chocs) > 0:
    favourite = random.choice(heroes)
    print("---")
    if show_favourite_chocolate(favourite=favourite,chocolates=chocs):
        chocs.remove(favourite)
````

````python
pulls = 0
number_of_sweets = 15
chocs = generate_random_sweets_list(number_of_sweets, distribution=heroes_distribution)
while len(chocs) > 0:
    favourite = random.choice(heroes)
    print("---")
    if show_favourite_chocolate(favourite=favourite,chocolates=chocs):
        chocs.remove(favourite)
    pulls += 1 
print(f"It took {pulls} pulls from the bag to get out all {number_of_sweets} sweets")
````
